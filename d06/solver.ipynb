```python
import numpy as np
content = open('data.txt').readlines()
s =''
for c in content:
    s+=c
s = s.split('\n\n');
groups = []
for x in s:
    groups.append(x.split('\n'))
print(len(groups))
t_sum = 0
for group in groups:
    all_in_group=''
    for person in group:
        all_in_group += person
    unique = ''.join(set(all_in_group))
    print(all_in_group)
    print(unique)
    n_unique = len(unique)
    t_sum+=n_unique;
    print("-----")
print(t_sum)
```

```python
k_sum = 0
for group in groups:
    group_list = []
    for person in group:
        group_list.append(person)
    if (len(group_list))==1:
        k= len(group_list[0])
        print("**********************")
    else:
        common = set(group_list[0]);
        for i in range(1,len(group_list)):
            common = common.intersection(group_list[i])
        k = len(list(common))
    k_sum +=k
    print([group_list,k,k_sum ])
    print("\n")
    
    
    
        
        
```
