```python
import numpy as np
# from collections import OrderedDict 


file1 = open('input.txt', 'r') 
Lines = file1.readlines() 
  
# Strips the newline character 

# Read conditions
cond={}
for i,line in enumerate(Lines): 
    a=line.strip().split(":")
    if len(a)==1:
        your_ticket = np.fromstring(Lines[i+2],sep=',',dtype=int)
        line_others = i+5;
        break
    cond[a[0]] = [ [int (a)for a in c.split('-')] for c in a[1].split(' or ')]
    
def is_valid(x,cond):
    for c in cond:
        for r in cond[c]:
            if x>=r[0] and x<=r[1]:
                return True
    return False

is_valid(43,cond)  

others=[]
for line in Lines[line_others:]:
    others.append(list(np.fromstring(line,sep=',',dtype=int)))

valid_tickets=[]
for o in others:
    v = [ is_valid(x,cond) for x in o]
    if all(v):
        valid_tickets.append(o)
valid_tickets = np.array(valid_tickets)

# print(your_ticket)
# print(valid_tickets)
# print(cond)
C = len(cond)
N = len(valid_tickets)
def is_ok(x,c):
    ranges = list(cond.values())[c] ;
    #print([ranges,x])
    for r in ranges:
        if x>=r[0] and x<=r[1]:
            return True
    return False

possible_list = []
for c0 in range(C):
    mat=np.zeros([C,N], dtype=bool)
    for c1 in range(C):
        for n in range(N):
            mat[c1,n] = is_ok(valid_tickets[n,c1], c0)
    #print(f"----------{list(cond.keys())[c0]}")
    #print(mat)
    #print(mat.all(axis=1))
    possible = (np.argwhere(mat.all(axis=1))).T[0]
    possible_list.append(list(possible))
    #print(possible)

    
    
print("-------------")
possible_dict = {}
for i,l in enumerate(possible_list):
    possible_dict[i] = l;
    print(f"{i} : {l}")
print("-------------")

solution={}
for loop in range(50):
    # Look for Unique
    for k in possible_dict.keys():
        if 1==len(possible_dict[k]):
            solved_col = k;
            allocated_cond = possible_dict[k][0];
            solution[solved_col] =  allocated_cond 
            print(f"Solved for Row {solved_col}, found answer {allocated_cond}")
            break;

    #Exclude found answer from other lists
    for k in possible_dict.keys():
        if allocated_cond in possible_dict[k]:
            possible_dict[k].remove(allocated_cond)
    possible_dict.pop(solved_col)
    if bool(possible_dict):
        pass;
    else:
        print("Solved")
        break;
print(solution)
target = list(cond.keys())[:6]
print(target)
print(your_ticket)
prod=1;
for t in range(6):
    print(your_ticket[solution[t]])
    prod*=your_ticket[solution[t]]
print(f"Answer : {prod}")
    
```
