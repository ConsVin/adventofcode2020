```python
import re
import numpy as np
file1 = open('inp_map.txt', 'r') 
Lines = file1.readlines() 
  
count = 0
m_valid=0;
# Strips the newline character 
mat=[]
for line in Lines: 
    s = line.replace(".", "0 ").replace("#","1 ")
    x = np.fromstring(s, dtype=int, sep=' ')
    mat.append(x)
mat=np.array(mat)

n_steps = mat.shape[0]
w= n_steps*3
k = int(np.ceil(w/ mat.shape[1]))
print(f"{n_steps*3}, mult {k}")
```

```python
print(mat.shape)
full_map = np.tile(mat,(1,k))
print(full_map.shape)
hits=0
for x in range(mat.shape[0]):
    y = x*3
    print([x,y,hits])
    hits += full_map[x,y]
print(hits)
```
