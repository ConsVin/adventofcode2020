```python
import numpy as np
fname = "input.txt"
lines = open(fname, "r").readlines()
for i,l in enumerate(lines):
    if l == "\n":
        break;
inputs= [line.strip() for line in lines[i+1:]]
print(inputs)


lines = open(fname, "r").readlines()
rules={}
for i,l in enumerate(lines):
    if l == "\n":
        break;
    s = l.strip().replace("\"","").split(": ")
    or_group = s[1].split(" | ")
    or_arr = []
    for and_group in or_group:
        arr = and_group.split(" ");
        and_str = "".join([f"(<{a}>)" for a in arr])
        or_arr.append(and_str)
    or_str = "|".join(or_arr)
        
    print(s[0],or_str)
    rules[s[0]] = or_str
if 1:
    rules['8'] = "(<42>)|(<42>)(<8>)"
    rules['11']= "(<42>)(<31>)|(<42>)(<11>)(<31>)"
                 
        
    
regex='<0>';
for loop in range(len(rules)):
    for k in rules.keys():
        regex = regex.replace(f"<{k}>",rules[k])
regex = regex.replace('((<a>))','a')
regex = regex.replace('((<b>))','b')

regex=regex.strip()+"\Z"
print(regex)

print(str(regex))
```

```python
import re
prog = re.compile(regex)
cnt=0;
for inp in inputs:
    x = prog.match(inp)
    if x:
        cnt+=1;
        print(f"Matched : {inp}");
    else:
        print(f"NOT Matched : {inp}");
        
print(cnt)
```
