```python
filename = 'input.txt'
with open(filename) as f:
    content = f.readlines()
commands = [x.strip().replace('+','').split(' ') for x in content] 

def run_commands(cmd_list):
    i = 0;
    acc=0
    visited=[]
    while True:
        if (i in visited):
            print("Infinite Loop found!");
            print(f"{i}:{cmd}:{arg}:{acc}")
            break;
        elif (i==len(commands)):
            print("*******************Program Terminates :)")
            print(f"{i}:{cmd}:{arg}:{acc}")
            break;
        else:
            visited.append(i)
        cmd,arg = cmd_list[i];
        arg = int(arg)
        if cmd=='acc':
            acc += arg;
            i+=1
        elif cmd=='jmp':
            i+=arg
        elif cmd=='nop':
            i+=1
#         print(f"{i}:{cmd}:{arg}:{acc}")
run_commands(commands)
```

```python
import copy
for i,c in enumerate(commands):
    cmd_modified = copy.deepcopy( commands);
    if (c[0]=='jmp'):
        print(f"Replace {i} jmp=>nop")
        cmd_modified[i][0]='nop'
        run_commands(cmd_modified)
    elif (c[0]=='nop'):
        print(f"Replace {i} nop=>jmp")
        cmd_modified[i][0]='jmp'
        run_commands(cmd_modified)
    else:
        print("Do noting for acc")
    
```
