```python
import numpy as np
content = open('boarding.txt').readlines()
prod_list=[]

for c in content:
    row = int(c[0:7].replace('F','0').replace('B','1'),2)
    col = int(c[7:10].replace('L','0').replace('R','1'),2)
    prod = row*8 + col
    prod_list.append(prod)
#     print([row,col,prod])
prod_list = np.array(prod_list)
prod_list = np.sort(prod_list)
for i,x in enumerate(prod_list[:-1]):
    print([i,x, x-i])
```
