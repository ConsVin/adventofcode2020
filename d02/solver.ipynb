```python
import re
file1 = open('input.txt', 'r') 
Lines = file1.readlines() 
  
count = 0
m_valid=0;
# Strips the newline character 
for line in Lines: 
    s = re.split(' |:|-|\n',line)
    n_min = int(s[0])
    n_max = int(s[1])
    letter =    s[2]
    passwrd = s[4]
    n =  passwrd.count(letter)
    if (n>= n_min) and ( n<=n_max):
        m_valid+=1;
    print("%d .. %d of %c -> %s %d %d"%(n_min, n_max, letter, passwrd, n, m_valid))
print(f"Valid passwords {m_valid}")    
    
```

```python
import re
file1 = open('input.txt', 'r') 
Lines = file1.readlines() 
  
count = 0
m_valid=0;
# Strips the newline character 

for line in Lines: 
    s = re.split(' |:|-|\n',line)
    n_first  = int(s[0])
    n_second = int(s[1])
    letter =    s[2]
    password = s[4]
    print(f"{n_first}..{n_second} {len(password)} {password}")
    
    check_second = password[n_first -1]==letter;
    check_first = password[n_second -1]==letter;
    ok = check_second ^ check_first
    print(f"{check_first}..{check_second}..{ok}")
    if ok:
        m_valid+=1

print(f"Valid passwords {m_valid}")    
    
    
```
