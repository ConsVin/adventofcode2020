```python
import numpy as np
l = open("input.txt", "r").readlines()
timestamp = int(l[0])

buses=[]
for t in l[1].split(","):
    if t!='x':
        buses.append(int(t))
timestamp
print(buses)
print(timestamp)

dep_times=[]
for b in buses:
    dep_times.append( np.ceil(timestamp/b)*b)
print(dep_times)
bus_id = buses[np.argmin(dep_times)]

wait_time = np.min(dep_times)- timestamp
ans = int(wait_time*bus_id)
print(f"Bus: {bus_id} at {np.min(dep_times)} wait time {wait_time}")
print(f"Ans:{ans}")

    
```
