```python
import numpy as np
fname='input.txt'
with open(fname, "r") as f:
    lines = f.readlines()
print(lines)

crs = 90;
pos = [0,0]
def sin(x):
    return np.sin(np.deg2rad(x))
def cos(x):
    return np.cos(np.deg2rad(x))

for cmd in lines:
    t = cmd[0]
    arg = int(cmd[1:].strip())
    if   t=='F':
        pos[ 1 ] += cos( crs )*arg;
        pos[ 0 ] += sin( crs )*arg;
    
    elif t=='L':
        crs = (crs-arg)%360
    elif t=='R':
        crs = (crs+arg)%360
    
    elif t=='N':
        pos[1]+= arg;
    elif t=='S':
        pos[1]-= arg;
        
    elif t=='E':
        pos[0]+= arg;
    elif t=='W':
        pos[0]-= arg;

ans = np.sum(np.abs(pos))
print(f"Ans:{ans}")
```

```python
import numpy as np
fname='input.txt'
with open(fname, "r") as f:
    lines = f.readlines()

course = [10,1]
pos    = [0,0]

def sin(x):
    return np.sin(np.deg2rad(x))
def cos(x):
    return np.cos(np.deg2rad(x))
def rotate(course, arg):
    x = course[0]
    y = course[1]
    course[0]=cos(arg)*x - sin(arg)*y
    course[1]=sin(arg)*x + cos(arg)*y
    return course 
        
for cmd in lines:
    t = cmd[0]
    arg = int(cmd[1:].strip())
    if   t=='F':
        pos[ 1 ] +=course[1] *arg;
        pos[ 0 ] +=course[0] *arg;
    elif t=='L':
        course = rotate(course, +arg)
    elif t=='R':
        course = rotate(course, -arg)
        
    elif t=='N':
        course[1]+= arg;
    elif t=='S':
        course[1]-= arg;
        
    elif t=='E':
        course[0]+= arg;
    elif t=='W':
        course[0]-= arg;
    print([cmd.strip(), np.round(course), np.round(pos)])

ans = np.sum(np.abs(pos))
print(f"Ans:{ans}")
```
